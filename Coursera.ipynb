# Неделя 2. Подбор k для KNN

```
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.preprocessing import scale
from sklearn.model_selection import cross_val_score, KFold
```

```
df = pd.read_csv("wine.data", header=None)
df
```

```
Y_df = df.iloc[:, 0]
Y_df
```

```
X_df = pd.DataFrame(df.iloc[:, 1:])
X_df
```

```
normalized_X_df = pd.DataFrame(scale(X_df))
normalized_X_df
```

```
folds = KFold(n_splits=5, shuffle=True, random_state=42)
k_to_accuracy = {}


for k in range(1, 51):
  knn = KNN(n_neighbors=k)
  k_to_accuracy[k] = cross_val_score(knn, X_df, Y_df, cv=folds, scoring='accuracy').mean()
```

```
normalized_k_to_accuracy = {}


for k in range(1, 51):
  knn = KNN(n_neighbors=k)
  normalized_k_to_accuracy[k] = cross_val_score(knn, normalized_X_df, Y_df, cv=folds, scoring='accuracy').mean()
```

```
for k in range(1,51):
  print('k = %d, accuracy = %f, normalized_accuracy = %f' % (k, k_to_accuracy[k], normalized_k_to_accuracy[k]))
```

```
def get_best_k (k_to_accuracy):
    max_accuracy = max((k_to_accuracy.values()))
    best_k = list(k_to_accuracy.keys())[list(k_to_accuracy.values()).index(max_accuracy)]
    print('Best k = %d, accuracy = %f' % (best_k, k_to_accuracy[best_k]))
    
```

```
get_best_k(k_to_accuracy)
get_best_k(normalized_k_to_accuracy)
```

# Неделя 2. Подбор метрик для KNN

```
from sklearn.neighbors import KNeighborsRegressor as KNR
from sklearn.datasets import load_boston
import numpy as np
```

```
data = load_boston()
X_df = pd.DataFrame(scale(data['data']))
X_df
```

```
Y_df = pd.DataFrame(data['target'])
Y_df
```

```
k_to_accuracy = {}
folds = KFold(n_splits=5, shuffle=True, random_state=42)

for p in np.linspace(1, 10, 200):
  knr = KNR(n_neighbors=5, weights='distance', metric='minkowski', p=p)
  k_to_accuracy[p] = cross_val_score(knr, X_df, Y_df, scoring='neg_mean_squared_error', cv=folds).mean()
k_to_accuracy
```

```
get_best_k(k_to_accuracy)
```

# Неделя 2. Перцептрон

```
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.linear_model import Perceptron
```

```
train_df = pd.read_csv("perceptron-train.csv", header=None)
train_df
```

```
test_df = pd.read_csv("perceptron-test.csv", header = None)
test_df
```

```
train_X = pd.DataFrame(train_df.iloc[:, 1:])
train_X
```

```
train_Y = train_df.iloc[:, 0]
train_Y
```

```
test_X = pd.DataFrame(test_df.iloc[:, 1:])
test_X
```

```
test_Y = test_df.iloc[:, 0]
test_Y
```

```
scaler = StandardScaler()
train_X_scaled = scaler.fit_transform(train_X)
test_X_scaled = scaler.transform(test_X)
```

```
not_scaled_clf = Perceptron()
not_scaled_clf = not_scaled_clf.fit(train_X, train_Y)

scaled_clf = Perceptron()
scaled_clf = scaled_clf.fit(train_X_scaled, train_Y)

not_scaled_accuracy = accuracy_score(test_Y, not_scaled_clf.predict(test_X))
print('Not scaled accuracy = %f' % not_scaled_accuracy)

scaled_accuracy = accuracy_score(test_Y, scaled_clf.predict(test_X_scaled))
print('Scaled accuracy = %f' % scaled_accuracy)

print('Difference = %f' % (scaled_accuracy - not_scaled_accuracy))
```
